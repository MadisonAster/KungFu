```python
fig, ax = plt.subplots(figsize=(10, 8))
# Set the axes through the origin
for spine in ['left', 'bottom']:
    ax.spines[spine].set_position('zero')
for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')
ax.set(xlim=(-5, 5), ylim=(-5, 5))
ax.grid()
vecs = ((2, 4), (-3, 3), (-4, -3.5))
for v in vecs:
    ax.annotate('', xy=v, xytext=(0, 0), arrowprops=dict(facecolor='blue', shrink=0, alpha=0.7, width=0.5)) 
                                                                                        #I think a longer line length is apropriate for notebooks
                                                                                        #Fewer lines = less scrolling to look at charts and equations
                                                                                        #Also equations tend to get long anyway, so I expect most people use a larger window
    ax.text(1.1 * v[0], 1.1 * v[1], str(v))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
# Set the axes through the origin
for spine in ['left', 'bottom']:
    ax.spines[spine].set_position('zero')
for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')

ax.set(xlim=(-5, 5), ylim=(-5, 5))
x = (2, 2)
ax.annotate('', xy=x, xytext=(0, 0),
            arrowprops=dict(facecolor='blue',
            shrink=0,
            alpha=1,
            width=0.5))
ax.text(x[0] + 0.4, x[1] - 0.2, '$x$', fontsize='16')


scalars = (-2, 2)
x = np.array(x)

for s in scalars:
    v = s * x
    ax.annotate('', xy=v, xytext=(0, 0),
                arrowprops=dict(facecolor='red',
                shrink=0,
                alpha=0.5,
                width=0.5))
    ax.text(v[0] + 0.4, v[1] - 0.2, f'${s} x$', fontsize='16')
plt.show()
```

Inner product of vectors $\large x,y∈\mathbb{R}^n$ is equal to:

$
\huge
x'y:=
\sum_{i=1}^n
x_i
y_i
$

The norm of a vector x represents its “length” (i.e., its distance from the zero vector) and is defined as

$
\huge∥
x
\huge∥
:=\sqrt{x'x}
:=
\begin{pmatrix}
    \sum_{i=1}^n x_i^2
\end{pmatrix}
$


```python
x = np.ones(3)            # Vector of three ones
y = np.array((2, 4, 6))   # Converts tuple (2, 4, 6) into array
print(x + y)
print(np.sum(x * y))          # Inner product of x and y
print(np.sqrt(np.sum(x**2)))  # Norm of x, take one
print(np.linalg.norm(x))      # Norm of x, take two
```
