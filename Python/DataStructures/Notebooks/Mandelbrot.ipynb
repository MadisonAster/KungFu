```python
import numpy 
import matplotlib.pyplot as pyplot

def mandelbrot(N_max, outer_threshold, nx, ny):
    x = numpy.linspace(-2, 1, nx)
    y = numpy.linspace(-1.5, 1.5, ny)

    c = x[:,numpy.newaxis] + 1j*y[numpy.newaxis,:]
    z = c
    for j in range(N_max):
        z = z**2 + c
    mandelbrot_set = (abs(z) < outer_threshold)
    return mandelbrot_set
mandelbrot_set = mandelbrot(50, 50., 601, 401)

pyplot.imshow(mandelbrot_set.T, extent=[-2, 1, -1.5, 1.5])
pyplot.gray()
pyplot.show()
```

Quadratic recurrance equation:

$
\huge
z_n+1=z+n^2+c
$

1-\sum_{n=1}^\infty n b_n^2
